```python
#This code is cited from Assignment 3 of cogs18
#check if the input has a question mark
def is_question(input_string):
    if '?' in input_string:
        output = True
    else:
        output = False
    return output
```

```python
#This code is cited from Assignment 3 of cogs18
#remove the punctuation of the input
def remove_punctuation(input_string):
    out_string=''
    for item in input_string:
        if item not in string.punctuation:
            out_string+=item
    return out_string
```

```python
#This code is cited from Assignment 3 of cogs18
#prepare the inputs for processing
def prepare_text(input_string):
    out_list=[]
    temp_string = input_string.lower()
    temp_string=remove_punctuation(temp_string)
    out_list=temp_string.split()
    return out_list
```

```python
#This code is cited from Assignment 3 of cogs18
#take a list of words that we got as input, a list of words to check for if 
#they appear in the input, and a list of possible outputs to return if something from 
#the list to check is in the input list
def selector(input_list,check_list,return_list):
    output = None
    if check_list == EMAIL_IN:
        for item in input_list:
            for i in check_list:
                if i in item:
                    output = random.choice(return_list)
                    break                    
    else:
        for item in input_list:
            if item in check_list:
                output = random.choice(return_list)
                break
    return output
```

```python
#This code is cited from Assignment 3 of cogs18
# concatenate two strings
def string_concatenator(string1,string2,separator):
    output=string1+separator+string2
    return output
```

```python
#This code is cited from Assignment 3 of cogs18
#transfer the elements in list to a string
def list_to_string(input_list,separator):
    output=''
    output=input_list[0]
    for item in input_list[1:]:
        output=string_concatenator(output,item,separator)
    return output
```

```python
#This code is cited from Assignment 3 of cogs18
#define how to exit the conversation
def end_chat(input_list):
    if 'quit' in input_list:
        output = True
    else:
        output = False
    return output
```

```python
#This code is cited from Assignment 3 of cogs18
def is_in_list(list_one, list_two):
    """Check if any element of list_one is in list_two."""
    
    for element in list_one:
        if element in list_two:
            return True
    return False

def find_in_list(list_one, list_two):
    """Find and return an element from list_one that is in list_two, or None otherwise."""
    
    for element in list_one:
        if element in list_two:
            return element
    return None
```

```python
#set inputs and outputs
GREETINGS_IN = ['hello', 'hi', 'hey', 'hola', 'welcome', 'bonjour', 'greetings']
GREETINGS_OUT = ['Hi, I am Olivia! I am your UCSD Online Evaluation Assistant \n\tCan you tell Olvia something about yourself? Are you a student,staff,or professor?']

CAPE_IN = ['cape']
CAPE_OUT = ['The old CAPE system has been substitued by me. Do you like Olivia?']

IDENTITY_IN_1 =['student']
IDENTITY_OUT_1 =['How about your finals! It must be stressful! What year are you? ']

IDENTITY_IN_2 = ['staff','professor']
IDENTITY_OUT_2 = ['Nice to meet you! what do you wish to tell Olivia?']

STUDENT_YEAR_IN= ['first','second','third','fourth','freshman','sophomore','junior','senior']
STUDENT_YEAR_OUT = ['Great!This is the best time of your college life.\n\tSo what do you wish to evaluate today?']

STUDENT_EVA_1_IN= ['parking','housing','dining','food','life','friend','parkings','resteruant','medical']
STUDENT_EVA_1_OUT =['I totally understand your feelings!.\n\tOn a scale of 0 to 5,0 being not emergent at all and 5 being extremely emergent, how would you rate your concerns?']
                    
RATE_LOW_IN = ['0','1','2','3']
RATE_LOW_OUT= ['Got it! So tell me how do you wish to improve the current situation?']

RATE_HIGH_IN = ['4','5']
RATE_HIGH_OUT = ['Olivia understand how emergent it is!\n\tCan you tell me about it and leave your email address so we can contact you to fix the problem?']
                
EMAIL_IN = ['com','edu'] # ###
EMAIL_OUT = ["Don't worry! Olivia will let someone contact you trough email as soon as possible!"]

STUDENT_EVA_2_IN = ['class', 'professor','ta','teaching','course']
STUDENT_EVA_2_OUT = ['So specifically which class do you want to evaluate']

PEOPLE_IN = ['cogs18', 'cogs1', 'econ1', 'mgt5','dsc10']
PEOPLE_OUT = ['is the professor of that course, right?']
PEOPLE_NAMES = {'cogs18': 'Ellis', 'cogs1': 'Boyle', 'ECON1': 'CANDIDO', 'MGT5': 'Housekeeper','DSC10':'Eldridge'}

ANSWER_IN =['yes']
ANSWER_OUT = ['So what do you wish to say about the course?']

PROBLEM_IN=['no']
PROBLEM_OUT = ["I'm sorry! so what do you wish to say about the course?"]
                    

COMMENT = ['Your responses have been recorded. Thank you!', 'Our staff will contact you soon to fix yout probelms!','Thank you for your feedbacks! Olivia will try to improve the situation!']
                 
QUESTION = "I'm sorry I don't have the answer for you question.\n\tBut you can email your question to oliviachat@ucsd.edu and my colleagues will answer the question for you"

```

```python
def olivia_chat():
    """Main function to run our chatbot."""
    
    display(Image(url='https://media.giphy.com/media/LHZyixOnHwDDy/giphy.gif'))
    
    
    chat = True
    while chat:

        # Get a message from the user
        msg = input('INPUT :\t')
        out_msg = None

        # Check if the input is a question
        question = is_question(msg)

        # Prepare the input message
        msg = prepare_text(msg)

        # Check for an end msg 
        if end_chat(msg):
            out_msg = 'Bye! Olivia hope to can see you again~'
            chat = False

        # Check for a selection of topics that we have defined to respond to
        #   Here, we will check for a series of topics that we have designed to answer to
        if not out_msg:
            
            
            # Initialize to collect a list of possible outputs
            outs = []

            # Check if the input looks like a greeting, add a greeting output if so
            outs.append(selector(msg,GREETINGS_IN,GREETINGS_OUT))

            # Check if people ask about the old CAPE system if so
            outs.append(selector(msg, CAPE_IN, CAPE_OUT))
            
            # Check if the visitors are students if so
            outs.append(selector(msg, IDENTITY_IN_1, IDENTITY_OUT_1))
            
            # Check if the visitors are professors and staffs if so
            outs.append(selector(msg, IDENTITY_IN_2, IDENTITY_OUT_2))
            
            # Check the year of the students  if so
            outs.append(selector(msg, STUDENT_YEAR_IN, STUDENT_YEAR_OUT))
            
            # Check if the student want to evaluate non-academic issues if so
            outs.append(selector(msg, STUDENT_EVA_1_IN, STUDENT_EVA_1_OUT))
            
            # Check if the situation is not emergent if so
            outs.append(selector(msg, RATE_LOW_IN, RATE_LOW_OUT))
            
            # Check if the situation is emergent if so
            outs.append(selector(msg, RATE_HIGH_IN, RATE_HIGH_OUT))
            
            # check if the situation is emergent and we have the email to give feedback if so
            outs.append(selector(msg, EMAIL_IN, EMAIL_OUT))
            
            # Check if the students want to evaluate acdemic issues if so
            outs.append(selector(msg, STUDENT_EVA_2_IN, STUDENT_EVA_2_OUT))
            
            # Check if we get the correct professor name for the course if so
            outs.append(selector(msg,ANSWER_IN, ANSWER_OUT))
            
            # Check if we get the correct professor name for the course if not so
            outs.append(selector(msg,PROBLEM_IN, PROBLEM_OUT))

            # Check if the input mentions a person that is specified, add a person output if so
            if is_in_list(msg, PEOPLE_IN):
                ipd.Audio(data = './UCSD.wav',autoplay = True)
                name = find_in_list(msg, PEOPLE_IN)
                outs.append(list_to_string([PEOPLE_NAMES[name],
                                            selector(msg, PEOPLE_IN, PEOPLE_OUT)], ' '))
            #display a gif to attract attentions
                display(Image(url='https://media.giphy.com/media/3ov9jVOiCsrNFHMeZO/giphy.gif'))

            
            options = list(filter(None,outs))
            if options:
                out_msg = random.choice(options)
                
        # If we don't have an output yet, but the input was a question, return msg related to it being a question
        if not out_msg and question:
            out_msg = QUESTION
        if not out_msg:
            out_msg = random.choice(COMMENT)


        print('OUTPUT:', out_msg)
```
